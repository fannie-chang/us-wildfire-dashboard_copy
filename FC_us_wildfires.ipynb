```python
# Find number of fire 

fires_year_size = pd.read_sql_query("""SELECT FIRE_NAME, FIRE_YEAR, STATE,DISCOVERY_DATE,CONT_DATE, FIRE_SIZE_CLASS, FIRE_SIZE, LATITUDE,LONGITUDE
                                 FROM fires 
                                 ORDER BY FIRE_YEAR DESC """
                                 ,engine)

fires_year_size
```
